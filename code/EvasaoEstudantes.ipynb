```python
#teste de cores para o dataframe
df = pd.DataFrame({
    'A': [1, -2, 4],
    'B': [-1, 3, -4],
})
def color_positive_negative(val):
    color = 'yellow'if val > 0 else 'red'
    return 'color: %s' % color

styled_df= df.style.applymap(color_positive_negative)
display(styled_df)
```

```python
#destaca o maximo e minimo 
def highlight_max_min(s):
    is_max = s == s.max()
    is_min = s == s.min()
    return ['color: yellow' if v_max else 'color:red' if v_min else '' for v_max, v_min in zip(is_max, is_min)]

styled_df = df.style.apply(highlight_max_min)
display(styled_df)
```

```python
pivot_df = stud.pivot_table(index='Marital status',
                            columns=['Target'],
                            values= 'Previous qualification (grade)',
                            aggfunc=np.mean
                            )
pivot_df
```

```python
#Destacar os valores maiores e menores de cada coluna. Como Ã© coluna se utilize apply()
def highlight_max_min(data):
    styles = data.copy()
    for col in data.columns: 
        max_val = data[col].max()
        min_val = data[col].min()
        styles[col] = ['background-color: blue' if v == max_val else 'background-color: yellow' 
                       if v == min_val else '' for v in data[col]
                       ]
    return styles

styled_df = pivot_df.style.apply(highlight_max_min, axis = None)
display(styled_df)
```
