Vejamos o que nós temos aqui:

count: Mostra o número de observações (linhas) não-nulas em cada coluna do DataFrame.

mean: Calcula a média aritmética dos valores em cada coluna. É a soma de todos os valores dividida pelo 
número de valores.

std: Representa o desvio padrão, que mede a quantidade de variação ou dispersão de um conjunto de valores. Um desvio padrão baixo indica que os valores tendem a estar próximos da média; um desvio padrão alto indica que os valores estão espalhados por uma gama mais ampla de valores.

min: Mostra o valor mínimo em cada coluna do DataFrame.

25%: Representa o primeiro quartil (Q1) dos dados, ou seja, 25% dos dados são menores que esse valor. É calculado ordenando os dados e tomando o valor no índice que é 25% do caminho através da lista.

50%: Representa a mediana ou o segundo quartil (Q2) dos dados. Metade dos dados são menores que esse valor. É o valor no meio de um conjunto de dados ordenado.

75%: Representa o terceiro quartil (Q3) dos dados, ou seja, 75% dos dados são menores que esse valor. É o valor abaixo do qual se encontram 75% dos dados.

max: Mostra o valor máximo em cada coluna do DataFrame.

```python
#Para o homem, o BMI de 20 a 24,9 é considerado normal. Acima disso, é considerado acima do peso, abaixo é considerado abaixo do peso. Crie uma nova coluna, chamada "bmi_male" que receberá o texto "abaixo do normal" para BMI abaixo de 20, "normal" para BMI de 20 a 24,9 e "acima do normal" para BMI acima de 24,9.
df['bmi_male'] = np.where(
    df['bmi'] < 20,
    'abaixo do normal',
    np.where(
        (df['bmi'] >= 20) & (df['bmi'] <= 24.9),
        'normal',
        'acima do normal'
    )
)
df.head()
```

```python
#Se uma homem é fumante e ainda tem um BMI acima do considerado normal, é possível que isso seja um indicador negativo para a seguradora. Quantos homens estão nessa situação?
grouped_data = df.groupby(['sex', 'smoker', 'bmi_male']).size().reset_index(name='count')
resultado = grouped_data[(grouped_data['sex']=='male') & (grouped_data['smoker'] == 'yes') & (grouped_data['bmi_male'] == 'acima do normal')]
print(resultado)
```

```python
#Crie uma coluna chamada regiao_sudeste e marque com 1 para quando a observação pertencer a um destes estados e 0, caso contrário. Faça o mesmo para a região centro-oeste. Qual a diferença no total de incêndios dessas regiões? Qual região possui mais incêndios? 
data['regiao_sudeste'] = np.where(
    data['state'].isin(['São Paulo', 'Rio de Janeiro', 'Minas Gerais', 'Espírito Santo']),
    1,
    0
)       
data['regiao_centro_oeste'] = np.where(
    data['state'].isin(['Mato Grosso', 'Mato Grosso do Sul', 'Goiás', 'Tocantins']),
    1,
    0
)
data.head()
```

```python
data.loc[data['regiao_sudeste'] == 1, 'number'].sum()
```

```python
data.loc[data['regiao_centro_oeste'] == 1, 'number'].sum()
```

```python
#Diferença no total de incêndios dessas regiões
data.loc[data['regiao_sudeste'] == 1, 'number'].sum() - data.loc[data['regiao_centro_oeste'] == 1, 'number'].sum()
```
